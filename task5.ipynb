```python
import cv2
import matplotlib.pyplot as plt

# Загрузка изображения
image = cv2.imread('./image.jpg')
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
# Создание объекта каскада Хаара для обнаружения лиц
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

# Обнаружение лиц на изображении
faces = face_cascade.detectMultiScale(image_gray)

# Отрисовка прямоугольника вокруг каждого обнаруженного лица
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

# Отображение изображения с нарисованными прямоугольниками вокруг лиц
cv2.imwrite("image-res.jpg", image)
faces_detected = cv2.imread("image-res.jpg")
faces_detected = cv2.cvtColor(faces_detected, cv2.COLOR_BGR2RGB)
plt.imshow(faces_detected)
```

```python
faces = face_cascade.detectMultiScale(image_gray, scaleFactor=1.3, minNeighbors=2)

for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

cv2.imwrite("image-res.jpg", image)
faces_detected = cv2.imread("image-res.jpg")
faces_detected = cv2.cvtColor(faces_detected, cv2.COLOR_BGR2RGB)
plt.imshow(faces_detected)
```
